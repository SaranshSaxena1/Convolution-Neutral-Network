```python
import scipy.io
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split

# Load University of Pavia dataset
def load_pavia():
    data = scipy.io.loadmat('PaviaU.mat')['paviaU']  # Hyperspectral data
    labels = scipy.io.loadmat('PaviaU_gt.mat')['paviaU_gt']  # Ground truth labels
    return data, labels

# Load Salinas dataset
def load_salinas():
    data = scipy.io.loadmat('Salinas.mat')['salinas']  # Hyperspectral data
    labels = scipy.io.loadmat('Salinas_gt.mat')['salinas_gt']  # Ground truth labels
    return data, labels

# Hyperspectral dataset class with patch extraction
class HyperspectralDataset(Dataset):
    def __init__(self, data, labels, patch_size=5, indices=None):
        self.data = data
        self.labels = labels
        self.patch_size = patch_size
        self.padding = patch_size // 2

        # Normalize data
        self.data = (self.data - np.min(self.data)) / (np.max(self.data) - np.min(self.data))

        # Pad the data
        self.padded_data = np.pad(self.data, ((self.padding, self.padding),
                                              (self.padding, self.padding),
                                              (0, 0)), mode='constant')

        # Use specified indices if provided, otherwise default to valid pixel indices
        if indices is None:
            self.indices = [(i, j) for i in range(self.labels.shape[0]) for j in range(self.labels.shape[1])
                            if self.labels[i, j] > 0]
        else:
            self.indices = indices

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        x, y = self.indices[idx]
        patch = self.padded_data[x:x + self.patch_size, y:y + self.patch_size]
        label = self.labels[x, y] - 1  # Convert labels to 0-based
        return torch.tensor(patch, dtype=torch.float32).permute(2, 0, 1), torch.tensor(label, dtype=torch.long)

# CNN model for hyperspectral data
class CNNModel(nn.Module):
    def __init__(self, input_channels, num_classes, patch_size):
        super(CNNModel, self).__init__()
        self.conv1 = nn.Conv2d(input_channels, 64, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(64)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(128)
        self.pool = nn.MaxPool2d(2)

        # Calculate the size of feature maps after convolutions and pooling
        dummy_input = torch.zeros(1, input_channels, patch_size, patch_size)
        dummy_output = self._forward_features(dummy_input)
        feature_map_size = dummy_output.view(-1).size(0)

        self.fc1 = nn.Linear(feature_map_size, 256)
        self.fc2 = nn.Linear(256, num_classes)

    def _forward_features(self, x):
        x = self.pool(nn.ReLU()(self.bn1(self.conv1(x))))
        x = self.pool(nn.ReLU()(self.bn2(self.conv2(x))))
        return x

    def forward(self, x):
        x = self._forward_features(x)
        x = x.view(x.size(0), -1)
        x = nn.ReLU()(self.fc1(x))
        x = self.fc2(x)
        return x

# Training function
def train_model(model, train_loader, criterion, optimizer, device, num_epochs=25):
    for epoch in range(num_epochs):
        model.train()
        total_loss = 0
        for data, labels in train_loader:
            data, labels = data.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(data)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1}/{num_epochs}, Loss: {total_loss/len(train_loader):.4f}")

# Testing function
def test_model(model, test_loader, device):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for data, labels in test_loader:
            data, labels = data.to(device), labels.to(device)
            outputs = model(data)
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)
    accuracy = correct / total
    print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Main script
if __name__ == "__main__":
    # Load data
    data, labels = load_pavia()  # Change to `load_salinas()` for Salinas dataset

    # Get all valid pixel indices (non-background)
    valid_indices = np.array([(i, j) for i in range(labels.shape[0])
                              for j in range(labels.shape[1]) if labels[i, j] > 0])

    # Split the valid indices into training and testing
    train_indices, test_indices = train_test_split(
        valid_indices, test_size=0.2, random_state=42,
        stratify=labels[valid_indices[:, 0], valid_indices[:, 1]]
    )

    # Prepare datasets
    patch_size = 5  # Define patch size
    train_dataset = HyperspectralDataset(data, labels, patch_size=patch_size, indices=train_indices)
    test_dataset = HyperspectralDataset(data, labels, patch_size=patch_size, indices=test_indices)

    # Create data loaders
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

    # Define model, loss, and optimizer
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    input_channels = data.shape[-1]  # Number of spectral bands
    num_classes = len(np.unique(labels)) - 1  # Ignore background (label 0)
    model = CNNModel(input_channels, num_classes, patch_size).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Train the model
    train_model(model, train_loader, criterion, optimizer, device, num_epochs=25)

    # Test the model
    test_model(model, test_loader, device)
```

```python
import scipy.io
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import pandas as pd
from sklearn.utils.class_weight import compute_class_weight

# Load datasets
def load_pavia():
    data = scipy.io.loadmat('PaviaU.mat')['paviaU']
    labels = scipy.io.loadmat('PaviaU_gt.mat')['paviaU_gt']
    return data, labels

def load_salinas():
    data = scipy.io.loadmat('Salinas.mat')['salinas']
    labels = scipy.io.loadmat('Salinas_gt.mat')['salinas_gt']
    return data, labels

# Dataset class with patch extraction
class HyperspectralDataset(torch.utils.data.Dataset):
    def __init__(self, data, labels, patch_size=5, indices=None):
        self.data = data
        self.labels = labels
        self.patch_size = patch_size
        self.padding = patch_size // 2

        # Normalize data
        self.data = (self.data - np.min(self.data)) / (np.max(self.data) - np.min(self.data))

        # Pad data
        self.padded_data = np.pad(self.data, ((self.padding, self.padding),
                                              (self.padding, self.padding),
                                              (0, 0)), mode='constant')

        # Use specified indices or default to valid pixel indices
        if indices is None:
            self.indices = [(i, j) for i in range(self.labels.shape[0]) for j in range(self.labels.shape[1])
                            if self.labels[i, j] > 0]
        else:
            self.indices = indices

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        x, y = self.indices[idx]
        patch = self.padded_data[x:x + self.patch_size, y:y + self.patch_size]
        label = self.labels[x, y] - 1  # Convert labels to 0-based
        return torch.tensor(patch, dtype=torch.float32).permute(2, 0, 1), torch.tensor(label, dtype=torch.long)

# CNN model for local and global features
class LocalGlobalCNNModel(nn.Module):
    def __init__(self, input_channels, num_classes, patch_size):
        super(LocalGlobalCNNModel, self).__init__()
        self.local_branch = nn.Sequential(
            nn.Conv2d(input_channels, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.global_branch = nn.Sequential(
            nn.Conv2d(input_channels, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        # Compute output sizes dynamically
        dummy_input = torch.zeros(1, input_channels, patch_size, patch_size)
        local_out_size = self._get_output_size(self.local_branch, dummy_input)
        global_out_size = self._get_output_size(self.global_branch, dummy_input)

        self.fc1 = nn.Linear(local_out_size + global_out_size, 256)
        self.fc2 = nn.Linear(256, num_classes)

    def _get_output_size(self, branch, dummy_input):
        with torch.no_grad():
            output = branch(dummy_input)
        return output.view(-1).size(0)

    def forward(self, x):
        local_features = self.local_branch(x)  # Get local features
        global_features = self.global_branch(x)  # Get global features

        local_features_flat = local_features.view(x.size(0), -1)  # Flatten local features
        global_features_flat = global_features.view(x.size(0), -1)  # Flatten global features
        combined_features = torch.cat((local_features_flat, global_features_flat), dim=1)

        x = nn.ReLU()(self.fc1(combined_features))
        x = self.fc2(x)
        return x

# Compute metrics
def compute_metrics(true_labels, pred_labels, num_classes):
    # Confusion Matrix
    cm = confusion_matrix(true_labels, pred_labels, labels=np.arange(num_classes))

    # Overall Accuracy (OA)
    OA = np.trace(cm) / np.sum(cm)

    # Average Accuracy (AA), handle cases where some classes are missing
    class_accuracies = []
    for i in range(num_classes):
        true_positive = cm[i, i]
        total = np.sum(cm[i, :])  # sum over row
        if total > 0:
            class_accuracies.append(true_positive / total)
    AA = np.mean(class_accuracies) if class_accuracies else 0

    # Kappa Coefficient
    total = np.sum(cm)
    P_o = np.trace(cm) / total
    P_e = np.sum(np.sum(cm, axis=0) * np.sum(cm, axis=1)) / (total ** 2)
    kappa = (P_o - P_e) / (1 - P_e)

    return OA, AA, kappa, cm


# Main function to visualize local features and compute metrics
def evaluate_model():
    # Load data
    data, labels = load_pavia()  # Change to `load_salinas()` for Salinas dataset

    # Get valid pixel indices (just a few for visualization)
    valid_indices = np.array([(i, j) for i in range(labels.shape[0])
                              for j in range(labels.shape[1]) if labels[i, j] > 0])

    # Split into training and testing
    train_indices, test_indices = valid_indices[:int(0.8 * len(valid_indices))], valid_indices[int(0.8 * len(valid_indices)):]

    # Prepare datasets
    patch_size = 5
    train_dataset = HyperspectralDataset(data, labels, patch_size=patch_size, indices=train_indices)
    test_dataset = HyperspectralDataset(data, labels, patch_size=patch_size, indices=test_indices)

    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True)
    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=False)

    # Define model
    input_channels = data.shape[-1]  # Number of spectral bands
    num_classes = len(np.unique(labels)) - 1  # Exclude background
    model = LocalGlobalCNNModel(input_channels, num_classes, patch_size)

    # Compute class weights for imbalanced dataset
    class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels.flatten())

    # If the class weights length is greater than num_classes, slice to match
    class_weights = class_weights[:num_classes]

    class_weights = torch.tensor(class_weights, dtype=torch.float32)
    criterion = nn.CrossEntropyLoss(weight=class_weights)

    # Define optimizer
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    # Training loop
    num_epochs = 10
    for epoch in range(num_epochs):
        model.train()
        for data, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(data)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

        print(f"Epoch {epoch+1}/{num_epochs}, Loss: {loss.item()}")

    # After training, evaluate the model
    model.eval()
    all_true_labels = []
    all_pred_labels = []

    with torch.no_grad():
        for data, labels in test_loader:
            outputs = model(data)
            _, predicted = torch.max(outputs, 1)

            all_true_labels.extend(labels.numpy())
            all_pred_labels.extend(predicted.numpy())

    # Compute metrics
    OA, AA, kappa, cm = compute_metrics(np.array(all_true_labels), np.array(all_pred_labels), num_classes)

    # Display results
    results = pd.DataFrame({
        'Overall Accuracy (OA)': [OA],
        'Average Accuracy (AA)': [AA],
        'Kappa Coefficient': [kappa]
    })

    print("Metrics:\n")
    print(results)
    print("\nConfusion Matrix:")
    print(cm)

# Run evaluation
evaluate_model()
```

```python
import scipy.io
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
from skfuzzy.cluster import cmeans
import pandas as pd


# Load datasets
def load_pavia():
    data = scipy.io.loadmat('PaviaU.mat')['paviaU']
    labels = scipy.io.loadmat('PaviaU_gt.mat')['paviaU_gt']
    return data, labels


# Apply fuzzy clustering
def apply_fuzzy_clustering(data, num_clusters=10):
    reshaped_data = data.reshape(-1, data.shape[-1])  # Flatten spatial dimensions
    reshaped_data = (reshaped_data - reshaped_data.min(axis=0)) / (reshaped_data.max(axis=0) - reshaped_data.min(axis=0))  # Normalize

    cntr, u, _, _, _, _, _ = cmeans(reshaped_data.T, num_clusters, 2, error=0.005, maxiter=1000)
    cluster_labels = np.argmax(u, axis=0)  # Assign data points to the cluster with highest membership probability
    return cluster_labels.reshape(data.shape[:2])


# Dataset class for fuzzy clustering
class HyperspectralDataset(torch.utils.data.Dataset):
    def __init__(self, data, labels, cluster_labels, patch_size=5):
        self.data = data
        self.labels = labels
        self.cluster_labels = cluster_labels
        self.patch_size = patch_size
        self.padding = patch_size // 2

        # Normalize data
        self.data = (self.data - np.min(self.data)) / (np.max(self.data) - np.min(self.data))

        # Pad data
        self.padded_data = np.pad(self.data, ((self.padding, self.padding),
                                              (self.padding, self.padding),
                                              (0, 0)), mode='constant')

        # Extract valid pixel indices
        self.indices = [(x, y) for x in range(labels.shape[0]) for y in range(labels.shape[1]) if labels[x, y] > 0]

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        x, y = self.indices[idx]
        patch = self.padded_data[x:x + self.patch_size, y:y + self.patch_size]
        label = self.labels[x, y] - 1  # Convert to 0-based labels
        cluster_label = self.cluster_labels[x, y]
        return torch.tensor(patch, dtype=torch.float32).permute(2, 0, 1), torch.tensor(label, dtype=torch.long), torch.tensor(cluster_label, dtype=torch.long)


# Improved CNN model with regularization
class ImprovedCNNModel(nn.Module):
    def __init__(self, input_channels, num_classes, patch_size):
        super(ImprovedCNNModel, self).__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(input_channels, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.fc = nn.Sequential(
            nn.Linear(128 * (patch_size // 4) ** 2, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x


# Compute metrics
def compute_metrics(true_labels, pred_labels, num_classes):
    cm = confusion_matrix(true_labels, pred_labels, labels=np.arange(num_classes))
    OA = np.trace(cm) / np.sum(cm)
    per_class_accuracy = np.diag(cm) / (np.sum(cm, axis=1) + 1e-10)
    AA = np.mean(per_class_accuracy)
    total = np.sum(cm)
    P_o = np.trace(cm) / total
    P_e = np.sum(np.sum(cm, axis=0) * np.sum(cm, axis=1)) / (total ** 2)
    kappa = (P_o - P_e) / (1 - P_e)
    return OA, AA, kappa, cm


# Training and evaluation function
def evaluate_model():
    # Load data
    data, labels = load_pavia()

    # Apply fuzzy clustering
    cluster_labels = apply_fuzzy_clustering(data, num_clusters=10)

    # Split into training and testing indices
    indices = [(x, y) for x in range(labels.shape[0]) for y in range(labels.shape[1]) if labels[x, y] > 0]
    np.random.shuffle(indices)
    train_indices = indices[:int(0.8 * len(indices))]
    test_indices = indices[int(0.8 * len(indices)):]

    # Create datasets
    patch_size = 5
    train_dataset = HyperspectralDataset(data, labels, cluster_labels, patch_size)
    test_dataset = HyperspectralDataset(data, labels, cluster_labels, patch_size)

    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True)
    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=False)

    # Model parameters
    input_channels = data.shape[-1]
    num_classes = len(np.unique(labels)) - 1

    # Class weights for imbalanced dataset
    class_weights = compute_class_weight('balanced', classes=np.arange(num_classes),
                                         y=[labels[x, y] - 1 for x, y in indices])
    class_weights = torch.tensor(class_weights, dtype=torch.float32)

    model = ImprovedCNNModel(input_channels, num_classes, patch_size)
    criterion = nn.CrossEntropyLoss(weight=class_weights)
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Training loop
    model.train()
    for epoch in range(10):
        total_loss = 0
        for data, target, _ in train_loader:
            optimizer.zero_grad()
            output = model(data)
            loss = criterion(output, target)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch + 1}/10, Loss: {total_loss / len(train_loader)}")

    # Testing loop
    model.eval()
    all_true_labels = []
    all_pred_labels = []
    with torch.no_grad():
        for data, target, _ in test_loader:
            output = model(data)
            _, predicted = torch.max(output, 1)
            all_true_labels.extend(target.numpy())
            all_pred_labels.extend(predicted.numpy())

    # Compute metrics
    OA, AA, kappa, cm = compute_metrics(np.array(all_true_labels), np.array(all_pred_labels), num_classes)

    # Display results
    results = pd.DataFrame({
        'Overall Accuracy (OA)': [OA],
        'Average Accuracy (AA)': [AA],
        'Kappa Coefficient': [kappa]
    })

    print("\nMetrics:\n", results)
    print("\nConfusion Matrix:\n", cm)


# Run evaluation
evaluate_model()
```
